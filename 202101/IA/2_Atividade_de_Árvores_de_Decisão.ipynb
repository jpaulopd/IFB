```python
url = 'http://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data' 

nome_var = ['comprimento_sepala', 'largura_sepala', 'comprimento_petala', 'largura_petala', 'especie']
iris = pd.read_csv(url, header=None, names=nome_var)

iris.head()
```

```python
iris.tail()
```

```python
iris['especie'] = iris['especie'].map({'Iris-versicolor': 0, 'Iris-setosa': 1, 'Iris-virginica':2})
iris
```

```python
# amostraVersicolor = (iris[iris['especie']=='Iris-versicolor'].sample(n=10, random_state=1))
# amostraSetosa = (iris[iris['especie']=='Iris-setosa'].sample(n=10, random_state=1))
# amostraVirginica = (iris[iris['especie']=='Iris-virginica'].sample(n=10, random_state=1))
amostraVersicolor = (iris[iris['especie']==0].sample(n=10, random_state=1))
amostraSetosa = (iris[iris['especie']==1].sample(n=10, random_state=1))
amostraVirginica = (iris[iris['especie']==2].sample(n=10, random_state=1))

amostraSetosa
```

```python
conjuntoteste = pd.concat([amostraVersicolor,amostraSetosa,amostraVirginica])
conjuntoteste
```

```python
conjuntotreinamento = iris.drop(conjuntoteste.index)
conjuntotreinamento
```

```python
classesVerdadeiras = conjuntoteste['especie'].tolist()

classesVerdadeiras
```

```python
# O seu código entra aqui
# Devemos obter um vetor de resposta chamado respostaArvore
from sklearn import tree
```

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(conjuntotreinamento[['comprimento_sepala','largura_sepala','comprimento_petala','largura_petala']],conjuntotreinamento[['especie']])
tree.plot_tree(clf) 
```

```python
respostaArvore = clf.predict(conjuntoteste[['comprimento_sepala','largura_sepala','comprimento_petala','largura_petala']])
respostaArvore
```

```python
errosArvore = 0

for i in range(len(conjuntoteste.index)):
  if classesVerdadeiras[i]!=respostaArvore[i]:
    errosArvore += 1

taxaErroArvore = errosArvore/len(conjuntoteste.index)

## Verificando o desempenho do classificador
print('Taxa de erro: ' + '%.2f' %(taxaErroArvore*100)+'% ou Acurácia de: ' + '%.2f' %((1-taxaErroArvore)*100)+'%')
```
