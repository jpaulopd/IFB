#3 - Verificar quantas ocorrências existem por estado.

```python
filtro_estado = mycol.aggregate(
   [
      {
         "$group":
                    {
                       "_id": "$estado", 
                       "count":{"$sum":1}
                    }
      },
      {
         "$sort":
                   {"count":-1}
      }
   ]
)

for i in filtro_estado:
  print(i)
```
